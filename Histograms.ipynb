```python
import pandas as pd

import plotly.graph_objects as go
import plotly.offline as pyo
```

```python
mpg = pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
data = go.Histogram(x=mpg.displacement)

layout = go.Layout(title='Distribution of Vehicle Displacement')

fig = go.Figure(data=data,layout=layout)

pyo.iplot(fig)
```

```python
data = go.Histogram(x=mpg.mpg,opacity=.5,xbins=dict(size=5),marker=dict(line=dict(color='black',width=1.5)))

layout = go.Layout(title='Distribution of Vehicle Miles Per Gallon')

fig = go.Figure(data=data,layout=layout)

pyo.iplot(fig)
```

### Histogram Exercise - Abalone data

```python
ab = pd.read_csv('data/abalone.csv')
ab.info()
```

```python
data = go.Histogram(x=ab['length'], xbins=dict(start=0, end=1,size=.02))
layout = go.Layout(title='Distribution of Abalone Length')
fig = go.Figure(data=data,layout=layout) 
pyo.iplot(fig)
```
